```python
# Aryan Bawa

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.table import Table

# reading the text file with the appropriate delimiters
whittaker_df = pd.read_csv('Tables/3dhst_whitaker14_fall2020.txt', header=0, sep='\\s+', comment='#')
whittaker_df.head()
```

```python
# specifying cols and rows
zmin = whittaker_df['zmin']
zmax = whittaker_df['zmax']
redshift = np.average([zmin, zmax], axis=0)

# making the plot
plt.scatter(whittaker_df['lmass'], whittaker_df['lSFR'], c = redshift)

# Adding some other stuff
plt.colorbar()
plt.grid()
plt.title('Whittaker 2014 - Stellar Mass vs SFR')
plt.xlabel('log(Stellar Mass)')
plt.ylabel('log(SFR)')
```
